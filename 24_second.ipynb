```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
basketball_data = pd.read_html("https://www.basketball-reference.com/leagues/NBA_stats.html")
```

```python
len(basketball_data)
```

```python
df = basketball_data[0]
df_clean = df.drop(df.index[[20, 21, 42, 43, 64, 65]])
df_clean.head()
```

```python
df_clean['PTS'] = pd.to_numeric(df_clean['PTS'])
df_clean['FG'] = pd.to_numeric(df_clean['FG'])
```

```python
points_24_second = df_clean.iloc[56:72, 22]
season_24_second = df_clean.iloc[56:72, 1]
```

```python
points_24_second
```

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.plot(season_24_second, points_24_second, 'go--', linewidth=2, markersize=12)
plt.title("Points Before and After 24 Second Shot Clock", fontsize=25)
plt.ylabel("Points", fontsize=20)
plt.xlabel("Season", fontsize=20)
plt.ylim((50,120))
plt.grid(True)
plt.vlines('1953-54', 0, 120)
plt.tick_params(labelsize = 12)
```

```python
FG_24_second = df_clean.iloc[56:72, 8]
FG_24_second
```

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.plot(season_24_second, FG_24_second, 'go--', linewidth=2, markersize=12)
plt.title("FGs Before and After 24 Second Shot Clock", fontsize=25)
plt.ylabel("FGs", fontsize=20)
plt.xlabel("Season", fontsize=20)
plt.ylim((0,50))
plt.grid(True)
plt.vlines('1953-54', 0, 120)
plt.tick_params(labelsize = 12)
```
