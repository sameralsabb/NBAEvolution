```python
import pandas as pd
import json 
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
filepath = os.path.join("leaguedashplayerbiostats.json")
with open(filepath) as jsonfile:
    nba_json = json.load(jsonfile)

nba_json    
```

```python
df = pd.DataFrame.from_dict(nba_json['resultSets'][0]['rowSet'], orient='columns')
table = df[[1, 3, 5, 6, 7, 13, 14, 15, 16]]
table.columns = ['Name', 'Team', 'Height', 'Height(in)', 'Weight', 'GP', 'Points', 'Rebounds', 'Assists']
table.head()
```

```python
NBA_2018 = table[table['GP'] >= 41]
NBA_2018.head()
```

```python
nba68 = pd.read_csv('nba1968.csv')
nba68.head()
```

```python
nba68[['Player']] = nba68.Player.str.split('\\', expand = True)[0]
nba68.drop_duplicates(subset=['Player'], keep=False)
nba1968 = nba68.drop(nba68.index[0])
nba68.head()
```

```python
nba1968 = nba68.drop(nba68.index[0])
```

```python
nba1968[['Ht','Ht(in)']] = nba1968['Ht'].str.split('-',expand=True)
nba1968[['Ht', 'Ht(in)']] = nba1968[['Ht', 'Ht(in)']].apply(pd.to_numeric, errors='coerce', axis=1)
nba1968['Ht'] = nba1968['Ht']*12
nba1968["Height(in)"] = nba1968["Ht"] + nba1968["Ht(in)"]
nba1968.rename(columns={"Pos" : "Position"}, inplace=True)
```

```python
Avg_1968 = nba1968.groupby(["Position"]).mean()
avgnba68 = Avg_1968[["Height(in)", "Wt"]]
print("These are the average Height and Weight per postion for the 1968 NBA/ABA Season")
print(avgnba68)
print("----------------------------------------------------")
print("These are the averages per player for the 1968 NBA/ABA Season")
print(nba1968.mean())
```

```python
rng = np.random.RandomState(0)
x = nba2018["Height(in)"]
y = nba2018["Weight"]
z = nba2018["Wins"]
colors = rng.rand(30)
plt.figure(figsize=(20,10), dpi=80)
plt.scatter(y, x, c=z, s=z*10, alpha=0.5, cmap='plasma')
plt.colorbar()
plt.title('Average Team Height/Weight vs Team Wins (2017-2018)', fontsize=25)
plt.ylabel('Height(in)', fontsize=20)
plt.xlabel('Weight', fontsize=20)
plt.xlim()
plt.grid()
```
