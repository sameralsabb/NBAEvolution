```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
basketball_data = pd.read_html("https://www.basketball-reference.com/leagues/NBA_stats.html")
```

```python
len(basketball_data)
```

```python
df = basketball_data[0]
df.columns = df.columns.droplevel()
df_clean = df.drop(df.index[[20, 21, 42, 43, 64, 65]])
df_clean.head()
```

```python
df_clean['PTS'] = pd.to_numeric(df_clean['PTS'])
df_clean['TOV'] = pd.to_numeric(df_clean['TOV'])
df_clean['3PA'] = pd.to_numeric(df_clean['3PA'])
```

```python
points_3_point_line = df_clean.iloc[29:47, 22]
season_3_point_line = df_clean.iloc[29:47, 1]
```

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.plot(season_3_point_line, points_3_point_line, 'go--', linewidth=2, markersize=12)
plt.title("Points Before and After 3-Point Line", fontsize=25)
plt.ylabel("Points", fontsize=20)
plt.xlabel("Season", fontsize=20)
plt.ylim(50,120)
plt.grid(True)
plt.vlines('1978-79', 50, 120)
plt.tick_params(labelsize = 12)
```

```python
Attempts_3_point_line = df_clean.iloc[23:48, 11]
season_3_point_line_Attempts = df_clean.iloc[23:48, 1]
```

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.plot(season_3_point_line_Attempts, Attempts_3_point_line, 'go--', linewidth=2, markersize=12)
plt.title("3PA After 3-Point Line Intoduction", fontsize=25)
plt.ylabel("3PA", fontsize=20)
plt.xlabel("Season", fontsize=20)
plt.ylim(0,16)
plt.grid(True)
plt.tick_params(labelsize = 12)
```

```python
tov_3_point_line = df_clean.iloc[28:50, 20]
season_3_point_line_tov = df_clean.iloc[28:50, 1]
```

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.plot(season_3_point_line_tov, tov_3_point_line, 'go--', linewidth=2, markersize=12)
plt.title("TOV Before and After 3-Point Line", fontsize=25)
plt.ylabel("Turnovers", fontsize=20)
plt.xlabel("Season", fontsize=20)
plt.ylim(15,22)
plt.grid(True)
plt.vlines('1978-79', 15, 22)
plt.tick_params(labelsize = 12)
```
